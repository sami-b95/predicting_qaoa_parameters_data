```python
from glob import glob
import matplotlib.pyplot as plt
from matplotlib import rc
import numpy as np
from pathlib import Path
import pickle

from lib import qaoa_erdos_renyi_normalize_parameters
```

# Configuration

```python
rc("font", size=18, weight="normal")
```

# Erdos-Renyi graphs

## Data loading

### Parameters

```python
DATADIR = Path("data", "infinite_size_erdos_renyi")
```

### Load all results

```python
results = {}
for result_idx, result_file in enumerate(glob(str(Path(DATADIR, "*", "result.pickle")))):
    print("result {}".format(result_idx))
    with open(result_file, "rb") as f:
        result = pickle.load(f)
    if result["d"] not in results:
        results[result["d"]] = []
    results[result["d"]].append(result)
```

## Optimal angles plots

### Parameters

```python
p = 2
```

### Extract optimal angles

```python
optimal_results = {
    d: min([result for result in results[d] if len(result["betas"]) == p], key=lambda result: result["energy"]["fun"], default=None)
    for d in results
    if d is not None
}
optimal_results = {d: optimal_result for d, optimal_result in optimal_results.items() if optimal_result}
d_values = np.array(sorted(list(optimal_results)))
optimal_betas = np.empty((len(d_values), p))
optimal_gammas = np.empty((len(d_values), p))
for d_idx, d in enumerate(d_values):
    result = optimal_results[d]
    optimal_betas_gammas = result["energy"]["x"]
    optimal_betas[d_idx, :], optimal_gammas[d_idx, :] = qaoa_erdos_renyi_normalize_parameters(2 * optimal_betas_gammas[:p], optimal_betas_gammas[p:])
    optimal_betas[d_idx, :] /= 2
```

### Plots

$\beta$ angles

```python
fig, ax = plt.subplots(figsize=(6, 4))
for beta_idx in range(optimal_betas.shape[1]):
    ax.plot(d_values, optimal_betas[:, beta_idx], "+-", label=f"$\\beta^*_{beta_idx}$")
ax.set_xlabel("$d$")
ax.set_ylabel("$\\beta^*$")
ax.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.4))
plt.savefig(Path("figures", "infinite_size", f"infinite_size_optimal_beta_p={p}.eps"), bbox_inches="tight")
```

$\gamma$ angles

```python
fig, ax = plt.subplots(figsize=(6, 4))
for gamma_idx in range(optimal_gammas.shape[1]):
    ax.plot(d_values, np.sqrt(d_values) * optimal_gammas[:, gamma_idx], "+-", label=f"$\gamma^*_{gamma_idx}$")
    ax.set_xlabel("$d$")
    ax.set_ylabel("$\gamma^*\sqrt{d}$")
ax.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.4))
plt.savefig(Path("figures", "infinite_size", f"infinite_size_optimal_gamma_p={p}.eps"), bbox_inches="tight")
```

# Chung-Lu graphs

## Data loading

### Parameters

```python
DATADIR = Path("data", "infinite_size_chung_lu")
```

### Load all results

```python
results = {}
for result_idx, result_file in enumerate(glob(str(Path(DATADIR, "*", "result.pickle")))):
    print("result {}".format(result_idx))
    with open(result_file, "rb") as f:
        result = pickle.load(f)
    key = tuple(result["ddist"])
    if key not in results:
        results[key] = []
    results[key].append(result)
```

## Optimal angles plots

### Parameters

```python
p = 5
degree_pair = (4, 9)
multi_betas = False
```

### Extract optimal angles

Optimal results per degree distribution

```python
optimal_results = {
    ddist: min(
        [
            result for result in results[ddist]
            if len(result["gammas"]) == p
            and ("multi_betas" not in result and multi_betas == False or "multi_betas" in result and result["multi_betas"] == multi_betas)
            and (result["ddist"][0][0], result["ddist"][1][0]) == degree_pair
            and result["ddist"][0][1] in np.linspace(0, 1, 10)
        ],
        key=lambda result: result["energy"]["fun"],
        default=None
    )
    for ddist in results
}
optimal_results = {ddist: optimal_result for ddist, optimal_result in optimal_results.items() if optimal_result}
```

Optimal angles for selected degree distribution

```python
degree_distributions = sorted(list(optimal_results), key=lambda degree_distribution: degree_distribution[1][1])
q_values = np.array([degree_distribution[1][1] for degree_distribution in degree_distributions])
if multi_betas:
    optimal_betas = np.empty((len(degree_distributions), len(degree_pair), p))
else:
    optimal_betas = np.empty((len(degree_distributions), p))
optimal_gammas = np.empty((len(degree_distributions), p))
for degree_distribution_idx, degree_distribution in enumerate(degree_distributions):
    result = optimal_results[degree_distribution]
    optimal_betas_gammas = result["energy"]["x"]
    if multi_betas:
        flat_betas, optimal_gammas[degree_distribution_idx, :] = qaoa_erdos_renyi_normalize_parameters(2 * optimal_betas_gammas[:len(degree_pair) * p], optimal_betas_gammas[len(degree_pair) * p:])
        optimal_betas[degree_distribution_idx, :, :] = flat_betas.reshape((p, len(degree_pair))).T
    else:
        optimal_betas[degree_distribution_idx, :], optimal_gammas[degree_distribution_idx, :] = qaoa_erdos_renyi_normalize_parameters(2 * optimal_betas_gammas[:p], optimal_betas_gammas[p:])
    optimal_betas[degree_distribution_idx, :] /= 2
```

### Plots

$\beta$ angles

```python
fig, ax = plt.subplots(figsize=(6, 4))
for beta_idx in range(p):
    if multi_betas:
        for beta_subidx in range(len(degree_pair)):
            ax.plot(q_values, optimal_betas[:, beta_subidx, beta_idx], "+-", label=f"$\\beta_{{{beta_idx}, {beta_subidx}}}$")
    else:
        ax.plot(q_values, optimal_betas[:, beta_idx], "+-", label=f"$\\beta^*_{beta_idx}$")
ax.set_xlabel("$q$")
ax.set_ylabel("$\\beta^*$")
ax.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.4))
plt.savefig(Path("figures", "infinite_size", f"infinite_size_optimal_beta_mixed_4_9_p={p}.eps"), bbox_inches="tight")
```

$\gamma$ angles

```python
fig, ax = plt.subplots(figsize=(6, 4))
avg_d = degree_pair[0] * (1 - q_values) + degree_pair[1] * q_values
for gamma_idx in range(optimal_gammas.shape[1]):
    ax.plot(q_values, optimal_gammas[:, gamma_idx] * np.sqrt(avg_d), "+-", label=f"$\gamma^*_{gamma_idx}$")
ax.set_xlabel("$q$")
ax.set_ylabel("$\gamma^*$")
ax.legend(loc="upper center", ncol=3, bbox_to_anchor=(0.5, 1.4))
plt.savefig(Path("figures", "infinite_size", f"infinite_size_optimal_gamma_mixed_4_9_p={p}.eps"), bbox_inches="tight")
```
